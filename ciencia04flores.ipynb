Vamos a averiguar que podemos hacer su tuvieramos valores nulos dentro d enuestros datos.
Para trabajar con estos valores se usa la libreria numpy

```python
import numpy as np
```

incluimos valores nulos en un par de filas: np.nan

```python
df.loc[:2, 'longitud_petalo'] = np.nan
df.head()
```

```python
df.isna().sum()
```

Los datos nulos son elementos de lo mas normal dentro de un DataFrame, datos qye bi se han rellenado. Para tratar datos nulos, tenemos un par de opciones
dropna() --> Elimina la columna/filas con nulos
Podemos indicar axis=0 para filas
axis = 1 para columnas

```python
df_limpio = df.dropna(axis=1)
df_limpio.head()
```

```python
df_filas = df.dropna(axis=0)
df_filas.head()
```

```python
df.loc[:4, 'longitud_petalo'] = np.nan
df.head(10)
```

La otra posibilidad está en poder operar con los datos dentro del DataFrame.
Un Null es lo mismo aquí que en base d edatos, no se tiene en cuenta.
En lugar d elimpiar el DataFrame de nuelos, podemos modificar valores nulos tambien.
Es decir, quitar el null y poner un valor por defecto.
Para ello se utiliza fillna(valor)

```python
df['longitud_petalo'] = df['longitud_petalo'].fillna(999)
df.head(10)
```

Tmbién podemos crear agrupaciones para averiguar caracteristicas respecto a una columna. esto es lo mismo que un group by de SQL

```python
df_grupo = df.groupby('especie')['ancho_petalo'].mean()
df_grupo
```
