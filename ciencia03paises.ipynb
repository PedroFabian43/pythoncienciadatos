Importamos pandas en el proyecto

```python
import pandas as pd
```

Cremaos un diccionario de paises

```python
dataPaises = {
    "pais": ["España", "Francia", "Colombia", "Mexico", "Perú", "Suiza"],
    "capital": ["Madrid", "París", "Bogotá", "CDMX", "Lima", "Praga"],
    "superficie": [1.9, 2.6, 3.9, 4.0, 2.8, 3.5]
}
```

Creamos un DataFrame con el diccionario.

```python
df = pd.DataFrame(dataPaises)
df
```

por ejemplo, podemos dibujar solamente una columna

```python
df[['capital', 'superficie']]
```

se pueden seleccionar filas por su posicion del índice.
Primero incluido y el segundo no incluido.

```python
df[2:5]
```

Por defecto, los indices d elos elementos son numéricos y los pone automaticamente.
podemos crear nuestros propios indices y sustituir el index


```python
df.index = ["ES", "FR", "CO", "MX", "PE", "SZ"]
df = df.reset_index()
df
```
